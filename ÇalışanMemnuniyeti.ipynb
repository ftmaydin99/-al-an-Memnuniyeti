```python
plt.figure(figsize=(10 , 6))
sns.scatterplot(data=df, x="Age", y="Salary", hue="Department")
plt.title("Ya Ve Maa likisi")
plt.xlabel("Ya")
plt.ylabel("Maa")
plt.show()
```

**alanlarn Cinsiyetine ve Departmanlarna Gre Ortalama Maa**

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Department", y="Salary", hue="Gender", data=df)
plt.title("Departman ve Cinsiyete Gre Ortalama Maa")
plt.xlabel("Departman")
plt.ylabel("Ortama Maa")
plt.xticks(rotation=45)
plt.show()
```

**Departman Baznda Memnuniyet**

```python
#Departman Baznda Ortalama Memnuniyet Oran Heasap
avg_satisfaction= df.groupby('Department')['Satisfaction Rate (%)'].mean()
```

```python
colors = {
    'Marketing':'blue',
    'IT':'orange',
    'Sales':'green',
    'HR':'red',
    'Finance':'purple'
}
```

```python
#ubuk Grafii
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_satisfaction.index, y= avg_satisfaction.values, palette=colors)
plt.title("Departman Baznda Ortalama Memnuniyet Oran")
plt.xlabel("Departman")
plt.ylabel("Ortalama Memnuniyet Oran(%)")
plt.show()
```

**Verimlilik ve Memnuniyet Oran likisi**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Satisfaction Rate (%)", y="Productivity (%)", hue="Department", s=100, alpha=0.7)
plt.title("Verimlilik ve Memnuniyet Oran Arasndaki liki")
plt.xlabel("Memnuniyet Oran (%)")
plt.ylabel("Verimlilik (%)")
plt.legend(title="Departman", bbox_to_anchor=(1,1) )
plt.show()
```

**Korelasyon Analizi**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = df[['Satisfaction Rate (%)', 'Productivity (%)']].corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Memnuniyet ve Verimlilik Arasndaki Korelasyon")
plt.show()
```

**Not!** Dk korelasyona ramen sreci tamamlamak adna basit bir regresyon modeli uyguladm.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
x = df[['Satisfaction Rate (%)']].values
y = df['Productivity (%)'].values
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='blue', label='Gerek Veriler')
plt.plot(x, y_pred, color='red', label='Regresyon Dorusu')
plt.title('Memnuniyet ve Verimlilik Arasndaki Basit Dorusal Regresyon')
plt.xlabel('Memnuniyet Oran (%)')
plt.ylabel('Verimlilik (%)')
plt.legend()
plt.show()
```

```python
print("Katsay (eim):", model.coef_[0])
print("Sabit (kesme):", model.intercept_)
```
