**Yaş Ve Maaş İlişkisi**

```python
plt.figure(figsize=(10 , 6))
sns.scatterplot(data=df, x="Age", y="Salary", hue="Department")
plt.title("Yaş Ve Maaş İlişkisi")
plt.xlabel("Yaş")
plt.ylabel("Maaş")
plt.show()
```

**Çalışanların Cinsiyetine ve Departmanlarına Göre Ortalama Maaşı**

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Department", y="Salary", hue="Gender", data=df)
plt.title("Departman ve Cinsiyete Göre Ortalama Maaş")
plt.xlabel("Departman")
plt.ylabel("Ortama Maaş")
plt.xticks(rotation=45)
plt.show()
```

**Departman Bazında Memnuniyet**

```python
#Departman Bazında Ortalama Memnuniyet Oranı Heasap
avg_satisfaction= df.groupby('Department')['Satisfaction Rate (%)'].mean()
```

```python
colors = {
    'Marketing':'blue',
    'IT':'orange',
    'Sales':'green',
    'HR':'red',
    'Finance':'purple'
}
```

```python
#Çubuk Grafiği
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_satisfaction.index, y= avg_satisfaction.values, palette=colors)
plt.title("Departman Bazında Ortalama Memnuniyet Oranı")
plt.xlabel("Departman")
plt.ylabel("Ortalama Memnuniyet Oranı(%)")
plt.show()
```

**Verimlilik ve Memnuniyet Oranı İlişkisi**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Satisfaction Rate (%)", y="Productivity (%)", hue="Department", s=100, alpha=0.7)
plt.title("Verimlilik ve Memnuniyet Oranı Arasındaki İlişki")
plt.xlabel("Memnuniyet Oranı (%)")
plt.ylabel("Verimlilik (%)")
plt.legend(title="Departman", bbox_to_anchor=(1,1) )
plt.show()
```

**Korelasyon Analizi**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = df[['Satisfaction Rate (%)', 'Productivity (%)']].corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Memnuniyet ve Verimlilik Arasındaki Korelasyon")
plt.show()
```

**Not!** Düşük korelasyona rağmen süreci tamamlamak adına basit bir regresyon modeli uyguladım.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
x = df[['Satisfaction Rate (%)']].values
y = df['Productivity (%)'].values
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='blue', label='Gerçek Veriler')
plt.plot(x, y_pred, color='red', label='Regresyon Doğrusu')
plt.title('Memnuniyet ve Verimlilik Arasındaki Basit Doğrusal Regresyon')
plt.xlabel('Memnuniyet Oranı (%)')
plt.ylabel('Verimlilik (%)')
plt.legend()
plt.show()
```

```python
print("Katsayı (eğim):", model.coef_[0])
print("Sabit (kesme):", model.intercept_)
```
